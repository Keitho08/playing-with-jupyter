 # Models
  

```python
def f(x, p): # p is the parameter that have been trained on a data set
    return p[0] + x * p[1]
```

## Analysis

```python
import seaborn as sns

# Load a dataset.
penguins = sns.load_dataset("penguins")
# Have a look at it.
sns.pairplot(penguins, hue="species")
```

```python
# Pick out two variables.
flipper = penguins[["body_mass_g", "flipper_length_mm"]].dropna()
# Scatter and fit line for just those two variables.
sns.regplot(x="body_mass_g", y="flipper_length_mm", data=penguins)
```

## Train

```python
import sklearn.linear_model as lin #sklearn is machine learning package..

x = flipper["body_mass_g"].to_numpy() 
y = flipper["flipper_length_mm"].to_numpy()

x = x.reshape(-1, 1)

model = lin.LinearRegression()
model.fit(x, y)
r = model.score(x, y) 
p = [model.intercept_, model.coef_[0]]
```

```python
r # measure how accurate line is on data set, 75% fit..
```

```python
p # if penguin is zero mody mass flipper lenght is 136 and lenght is 0.015mm, 
```

## Predict

```python
f(4500.0, p) # predicting flipper lenght based on body mass in grams is 0 (scales up per added gram)
```

```python
def predict(x):
    return f(x, p) # uses p value above to predict based on body mass input
```

```python
predict(4500.00) # flipper lenght of 205mm
```
